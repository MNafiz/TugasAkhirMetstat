```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
url_data = "https://raw.githubusercontent.com/MNafiz/TugasAkhirMetstat/master/dataset/datasetmetstat.csv"
df = pd.read_csv(url_data)

# masukkin data dari luar
```

```python
df.reset_index(drop=True)
```

```python
for i in range(34):
  df["UMP"][i] = df["UMP"][i].replace(",","")

# merubah UMP biar jadi tipe data numerik
```

```python
df["UMP"] = df['UMP'].astype(float)
# merubah UMP biar jadi tipe data numerik
```

```python
df.head()
```

```python
df.set_index("Provinsi")
```

```python
df.drop(["No"],axis=1,inplace=True)
```

```python
df.corr()

# Mengetahui seberapa kuat hubungan antara variabel UMP dan IPN pada tahun 2021
```

```python
plt.scatter(df.UMP,df.IPN)

#plot data data dengan sumbu x menggambarkan UMP dan sumbu Y menggambarkan IPN
```

```python
reg = np.polyfit(df.UMP,df.IPN,deg = 1)

# mendapatkan slope dan intercept (mendapatkan persamaan garis yang paling menggambarkan sebaran data)
```

```python
reg
```

```python
line = np.polyval(reg,df.UMP)
plt.scatter(df.UMP,df.IPN)
plt.plot(df.UMP,line,"r")
```
